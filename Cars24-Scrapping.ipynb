```python
import re
import warnings
warnings.filterwarnings("ignore")
import requests
import numpy as np
from bs4 import BeautifulSoup

Year=[] 
Brand=[] 
Model=[] 
Fuel_Type=[] 
Transmission=[]
KM_Driven=[] 
Price=[] 
EMI_Price=[] 
Location =[]

data = {
    "pune": 2423, "bangalore": 4709, "mumbai": 2378, "delhi": 2, "hyderabad": 3686, "nashik": 2598,
    "indore": 2920, "noida": 134, "chandigarh": 8623, "ludhiana": 666, "lucknow": 290, "gurgaon": 5,
    "ghaziabad": 132, "kolkata": 777, "coimbatore": 6105, "patna": 8184, "surat": 1605, "vadodara": 1674,
    "agra": 136, "ahmedabad": 1692, "jaipur": 2130, "amritsar": 684, "trivandrum": 6565,
    "bhubaneswar": 7323, "dehradun": 1512, "bhopal": 2987
}

for city, city_id in data.items():
    u = f"https://www.cars24.com/buy-used-cars-{city}/?sort=bestmatch&serveWarrantyCount=true&storeCityId={city_id}"
    request_header = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    }
    page = requests.get(u, headers=request_header)
    soup = BeautifulSoup(page.text, "html.parser")

    cars = soup.find_all('div', class_='styles_contentWrap__9oSrl')

    for car in cars:
        text_block = car.find("div", class_="styles_outer__NTVth")
        text = text_block.text if text_block else ""

        
        year = re.findall(r'^\d+', text)
        Year.append(year[0] if year else np.nan)

        
        brand = re.findall(r'^\d+\s(\w+)', text)
        Brand.append(brand[0] if brand else np.nan)

        
        model = re.findall(r'\d{4}\s[a-zA-Z]+\s(\w+)', text)
        Model.append(model[0] if model else np.nan)

      
        fuel = re.findall(r'(Petrol|Diesel|CNG)', text)
        Fuel_Type.append(fuel[0] if fuel else np.nan)

        
        transmission = re.findall(r'(Manual|Auto)', text)
        Transmission.append(transmission[0] if transmission else np.nan)

        
        km = re.findall(r'(\d+\.?\d*\s?\w+ km)', text)
        KM_Driven.append(km[0] if km else np.nan)


        price = re.findall(r'([\d.]+)(?=\s*lakh)', text)
        Price.append(price[0] if price else np.nan)

        emi_price = re.findall(r'EMI \D\d+,\d+\/m', text)
        EMI_Price.append(emi_price[0] if emi_price else np.nan)

        loc_tag = car.find("div", class_="styles_ellipsis__uatjG")
        Location.append(loc_tag.text.strip() if loc_tag else np.nan)

```

```python
# Extract actual price and EMI price as shown on Cars24 website for a given city
import requests
from bs4 import BeautifulSoup
import re

city = 'pune'  # Change city as needed
city_id = 2423  # Change city_id as needed
url = f'https://www.cars24.com/buy-used-cars-{city}/?sort=bestmatch&serveWarrantyCount=true&storeCityId={city_id}'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html.parser')
cars = soup.find_all('div', class_='styles_contentWrap__9oSrl')

for car in cars:
    text_block = car.find('div', class_='styles_outer__NTVth')
    text = text_block.text if text_block else ''

    # Extract actual price
    price_match = re.search(r'₹[\d,]+(?:\.\d+)?\s?(?:lakh|L)?', text)
    price = price_match.group(0) if price_match else 'N/A'

    # Extract EMI price
    emi_match = re.search(r'EMI [^\d]*(\d+,\d+)/m', text)
    emi = emi_match.group(1) if emi_match else 'N/A'

    print(f'Actual Price: {price} | EMI Price: {emi}')
```
